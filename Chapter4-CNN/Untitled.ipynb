```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense 
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Tao khung chua
classifier = Sequential()
#Step 1 convolution
classifier.add(Conv2D(32,(3,3), input_shape=(64,64,3), activation='relu'))
#gom 32 bo loc, 1 bo shape 3,3, anh dau vao 64X64 va co mau, ham kich hoat la relu
```

```python
#step 2 - max pooling
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#add a second convolution layer
classifier.add(Conv2D(32,(3,3), activation='relu'))#lop conv thu 2 khong can phai co input shape
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#step 3 - flattening
classifier.add(Flatten())
```

```python
#Step 4 ANN: full connection
classifier.add(Dense(units = 128, activation = 'relu'))#hidden node canh theo hinh
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
#compilling the CNN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen  = ImageDataGenerator(rescale= 1./255, #scale gia tri diem anh
                                   shear_range = 0.2, # xoay 20%
                                   zoom_range = 0.2, #phong 20%
                                   horizontal_flip = True) #xoay hinh
```

```python
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/LDS8_Le Quang Anh Tu/week2/Chapter4/dataset/training_set',
                                                target_size = (64,64),
                                                batch_size = 32,
                                                class_mode = 'binary')
```

```python
test_set = train_datagen.flow_from_directory('/content/drive/MyDrive/LDS8_Le Quang Anh Tu/week2/Chapter4/dataset/test_set',
                                                target_size = (64,64),
                                                batch_size = 32,
                                                class_mode = 'binary')
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint #early de dung model som khi cac gia tri it thay doi
#modelcheckpoint dung cho viec khi dang chay co truc trac xay ra, khi chay lai o checkpoint
```

```python
callbacks = [EarlyStopping(monitor='val_loss', patience=5), #dung gia tri monitor = val_loss de canh, 5 lan giong nhau thi ngung lai
            ModelCheckpoint('/content/drive/MyDrive/LDS8_Le Quang Anh Tu/week2/Chapter4/model_checkpoint_new.h5', save_best_only=True)] #duoi modelckp la h5, duoi sklearn la .pkl

#reload model neu gap loi chua xong
#f from tensorflow.keras.models import load_model
# classifier = load_model('model_checkpoint_new.h5')
```

```python
classifier.fit(training_set,
              steps_per_epoch = len(training_set),
              epochs=100,
              validation_data = test_set,
              validation_steps = len(test_set),
              callbacks = callbacks)
```
